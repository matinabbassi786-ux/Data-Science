```python
with open("HTML/page-1.html","r") as f:
    rowhtml = f.read()
    
soup = BeautifulSoup(rowhtml, "lxml")
```

```python
h3s = soup.find_all('h3')
```

```python
articles = soup.select("article.product_pod")
```
