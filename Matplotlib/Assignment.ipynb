```python
import matplotlib.pyplot as plt
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
years
```

```python
Kohli = [100,142,190,250,200,141,111,100,75,30]
Kohli
```
